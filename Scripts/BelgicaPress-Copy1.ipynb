```python
annot = pd.read_csv(r'06_QueriesByVisit_towork2.csv', encoding="utf-8", engine="c", index_col="unique_id")

annot = annot.drop("query_tokens", axis=1)
annot = annot.rename(columns={'query_normalised': 'queries'})
annot.head()
```

```python
prenoms = pd.read_csv(r"C:\Users\Boulot\Desktop\BelgicaPress\prenoms.tsv", encoding="utf8")
prenoms = set([str(name).strip().lower() for name in prenoms.prenoms])
```

```python
noms_famille = pd.read_csv(r"C:\Users\Boulot\Desktop\BelgicaPress\Names.csv", encoding="utf8", error_bad_lines=False)
noms_famille = set([name.strip().lower() for name in noms_famille.names])
```

```python
communes_complet = pd.read_csv(r"C:\Users\Boulot\Desktop\BelgicaPress\communes_et_sections.csv", encoding="utf8")
```

```python
liste_mots = pd.read_csv(r"C:\Users\Boulot\Desktop\BelgicaPress\liste_mots.tsv", encoding="utf8")
liste_mots = set([name.strip().lower() for name in liste_mots.mots])
```

```python
geo = pd.read_csv(r"C:\Users\Boulot\Desktop\BelgicaPress\communes_et_sections_noms_unifies.csv", encoding="utf8")
geo = geo.drop_duplicates(['noms_unifie'])
geo['lower'] = geo['noms_unifie'].str.lower()
lieux_be = set([str(name).strip() for name in geo.lower])
```
